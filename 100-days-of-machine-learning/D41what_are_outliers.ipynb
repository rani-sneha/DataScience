#What are Outliers?

Outliers are those set of data which don't fit with normal data.

Eg:- if we draw a graph where

x: marks of student

y: no. of hours of study

There would be some students whose marks would be high despite them giving less hours of study and some students whose marks would be low even if they have studied. These students are outliers.

So outliers can affect our data in both positive and negative ways. We have to decide and treat accordingly. Likewise:-

*Positive Impact* :- If you are searching why there is an anamoly in the ATM machine and if you remove the outliers you won't be able to find the anamoly.

*Negative Impact*:- For the above case of hours of study and marks if we don't remove the outliers and apply linear regression then our model will not work effectively as desired.


**Algorithms affected by outliers?**

All those algorithms where weight is taken into consideration is affected by outliers. Like:- Linear regression, Logistic regression, Adabost, Deep Learning etc.

Not Affected:-
  Tree Based algorithms.


#How to remove outliers?

There are 4 ways in which outliers are treated. They are:-

1. Trimming - Remove the outliers

  Advantage:-

  a. Fast

  b. Easy

  Disadvantage:-

  a. if outlier present in large number, it can thin the data

2. Capping -  In this method, we mostly set a limit that above this or below this whatever points we have are outliers and we cap them.
Eg:-

if 5 is lower limit and 80 is upper limit and we get values like 3 and 120 then 3 would be set to 5 and 120 will be set to 80.

3. Treating as Missing values
4. Discritaization - You sepreate the values into buckets and the outliers get into the first and last bucket.
